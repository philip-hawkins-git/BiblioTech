**BiblioTech** is a simple, user-friendly script designed to format book titles according to either the American Psychological Association, Associated Press, Chicago Manual Style, or Modern Language Association style guides. Simply enter the text and run the script and let BiblioTech do what it does best. Enjoy.

```python
#@title BiblioTech: A Book Title Formatting Script

BOOK_TITLE = "" # @param {type:"string"}
STYLE = "American Psychological Association" # @param ["American Psychological Association", "Associated Press", "Chicago Manual of Style", "Modern Language Association"]

import re

if STYLE == "American Psychological Association":
    def apa_title_case(text):
        # Define words that should always be capitalized.
        always_capitalize = set([
            "a", "an", "the", "and", "but", "or", "nor", "at", "by", "for", "in", "of", "on", "to",
        ])

        # Split the input text into words.
        words = re.findall(r'\b\w+\b', text)

        # Capitalize the first word, the last word, and major words.
        for i in range(len(words)):
            word = words[i]
            # Check if the word is not in the always_capitalize set.
            if word.lower() not in always_capitalize:
                words[i] = word.capitalize()

        # Reconstruct the title with the formatted words.
        formatted_title = ' '.join(words)

        return formatted_title

    # Format the book title using APA style capitalization rules
    formatted_title = apa_title_case(BOOK_TITLE)

    # Print the formatted title
    print("American Psychological Association Format:")
    print(formatted_title)


if STYLE == "Associated Press":
    def ap_title_case(text):
        # Define words to be excluded from capitalization (articles, prepositions, and conjunctions).
        exclude_words = set([
            "a", "an", "the", "and", "but", "or", "nor", "at", "by", "for", "in", "of", "on", "to", "up", "as", "it"
        ])

        # Split the input text into words.
        words = re.findall(r'\b\w+\b', text)

        # Capitalize the first word regardless of its nature.
        words[0] = words[0].capitalize()

        # Capitalize subsequent words if they are not in the exclude_words list.
        for i in range(1, len(words)):
            if words[i].lower() not in exclude_words:
                words[i] = words[i].capitalize()

        # Reconstruct the title with the formatted words.
        formatted_title = ' '.join(words)

        return formatted_title

    # Format the book title using AP style book title capitalization rules
    formatted_title = ap_title_case(BOOK_TITLE)

    # Print the formatted title
    print("Associated Press Format:")
    print(formatted_title)

if STYLE == "Chicago Manual of Style Format":
    def chicago_title_case(text):
        # Define words that should always be capitalized.
        always_capitalize = set([
            "a", "an", "the", "and", "but", "or", "nor", "at", "by", "for", "in", "of", "on", "to", "up",
        ])

        # Split the input text into words.
        words = re.findall(r'\b\w+\b', text)

        # Capitalize the first word and the last word.
        if len(words) > 0:
            words[0] = words[0].capitalize()
        if len(words) > 1:
            words[-1] = words[-1].capitalize()

        # Capitalize major words and words in hyphenated compounds.
        for i in range(1, len(words) - 1):
            word = words[i]
            # Check if the word is not in the always_capitalize set and not "with."
            if word.lower() not in always_capitalize and word.lower() != "with":
                words[i] = word.capitalize()

        # Reconstruct the title with the formatted words.
        formatted_title = ' '.join(words)

        return formatted_title

    # Format the book title using Chicago Manual of Style capitalization rules
    formatted_title = chicago_title_case(BOOK_TITLE)

    # Print the formatted title
    print("Chicago Manual of Style Format:")
    print(formatted_title)

if STYLE == "Modern Language Association":
    def mla_title_case(text):
        # Words that should be lowercase in titles
        lowercase_words = [
            "a", "an", "and", "as", "at", "but", "by", "for", "from", "in",
            "nor", "of", "on", "or", "over", "the", "to", "up", "with"
        ]

        # Split the input text into words
        words = re.findall(r'\w+', text)

        # Capitalize the first word
        words[0] = words[0].capitalize()

        # Capitalize subsequent words unless they are in lowercase_words
        for i in range(1, len(words)):
            if words[i].lower() not in lowercase_words:
                words[i] = words[i].capitalize()

        # Reconstruct the title
        formatted_title = ' '.join(words)

        return formatted_title

    formatted_title = mla_title_case(BOOK_TITLE)
    print("Modern Language Association Format:")
    print(formatted_title)
```
